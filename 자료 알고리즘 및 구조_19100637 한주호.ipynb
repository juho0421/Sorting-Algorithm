## 1. Bubble Sort

```python
# 서로 이웃한 원소들을 순서대로 비교하여 정렬
def bubbleSort(arr1):
        n = len(arr1)
            
        for i in range(n):

            for j in range(0, n-i-1):

                if arr1[j] > arr1[j+1]:
                    arr1[j], arr1[j + 1] = arr1[j+1], arr1[j]


arr1 = []

with open(name_path, 'r') as file:
    line = None
    while line != '':
        line = file.readline()
        arr1.append(line.strip('\n'))

Time = time.time()
bubbleSort(arr1)
Total_Time1 = time.time() - Time

print("[Sorted array is]")
for i in range(len(arr1)):
    print("[%04d] %s " %(i, arr1[i]), end = "\n")

print(Total_Time1)
```

## 2. Insertion Sort

```python
# i번째 요소(key)를 i-1부터 1까지의 요소와 비교하고 삽입 후 한칸씩 이동
def insertionSort(arr2):
    n = len(arr2)

    for i in range(1, n):
        key = arr2[i]
        j = i-1
        while j >= 0 and key < arr2[j]:
            arr2[j+1] = arr2[j]
            j -= 1
        arr2[j+1] = key


arr2 = []

with open(name_path, 'r') as file:
    line = None
    while line != '':
        line = file.readline()
        arr2.append(line.strip('\n'))

Time = time.time()
insertionSort(arr2)
Total_Time2 = time.time() - Time

print("[Sorted array is]")
for i in range(len(arr2)):
    print("[%04d] %s " %(i, arr2[i]), end = "\n")

print(Total_Time2)
```

## 3. Selection Sort

```python
# 모든 요소들을 확인 후 가장 작은 요소를 앞으로 이동 반복
def selectionSort(arr3):
    n = len(arr3)

    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr3[j] < arr3[min_idx]:
                min_idx = j                           # 최소값의 index 찾기
        arr3[i], arr3[min_idx] = arr3[min_idx], arr3[i]


arr3 = []



with open(name_path, 'r') as file:
    line = None
    while line != '':
        line = file.readline()
        arr3.append(line.strip('\n'))

Time = time.time()
selectionSort(arr3)
Total_Time3 = time.time() - Time

print("[Sorted array is]")
for i in range(len(arr3)):
    print("[%04d] %s " %(i, arr3[i]), end = "\n")

print(Total_Time3)
```

## 4. Merge Sort

```python
# 원소 개수가 1 또는 0이 될 때까지 집단 L, 집단 R으로 쪼개고, 쪼개진 요소끼리 비교한 뒤 병합을 반복 
# => L의 요소와 R의 요소의 0번 인덱스의 데이터를 비교한 뒤, 작은 값이 나온 집단의 인덱스를 증가
# O(n log n)
def mergeSort(arr4):
    if len(arr4) > 1:
        mid = len(arr4)//2
        L = arr4[:mid]
        R = arr4[mid:]

        mergeSort(L)
        mergeSort(R)

        i = j = k = 0

        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr4[k] = L[i]
                i += 1
            else:
                arr4[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            arr4[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr4[k] = R[j]
            j += 1
            k += 1


arr4 = []

with open(name_path, 'r') as file:
    line = None
    while line != '':
        line = file.readline()
        arr4.append(line.strip('\n'))
        
Time = time.time()
mergeSort(arr4)
Total_Time4 = time.time() - Time

print("[Sorted array is]")
for i in range(len(arr4)):
    print("[%04d] %s " %(i, arr4[i]), end = "\n")

print(Total_Time4)
```

## 5. Quick Sort

```python
# 가장 마지막 원소를 pivot으로 설정, pivot보다 작은 것은 전부 앞쪽으로 이동, 이동이 끝나면 중간 인덱스의 원소를 pivot으로 재지정, 반복
# O(n log n) but 최악의 경우 O(n²)
def quickSort(arr5, low, high): # low와 high가 같아질 때까지 반복
    if low < high:
        pivot = partition(arr5, low, high)

        quickSort(arr5, low, pivot - 1)
        quickSort(arr5, pivot + 1, high)

def partition(arr5, low, high): # low부터 high까지의 arr의 요소를 정렬
    pivot = arr5[high]
    i = low 

    for j in range(low, high):
        if arr5[j] < pivot:
            arr5[i], arr5[j] = arr5[j], arr5[i]
            i += 1

    arr5[i], arr5[high] = arr5[high], arr5[i]
    return i 

arr5 = []

with open(name_path, 'r') as file:
    line = None
    while line != '':
        line = file.readline()
        arr5.append(line.strip('\n'))

Time = time.time()
n = len(arr5)
quickSort(arr5, 0, n-1)
Total_Time5 = time.time() - Time

print("[Sorted array is]")
for i in range(len(arr5)):
    print("[%04d] %s " %(i, arr5[i]), end = "\n")

print(Total_Time5)
```

## 6. Bucket Sort 

```python
# 크기가 n인 bucket을 만듦, 각 bucket에 해당하는 원소 삽입, bucket 안의 원소들 정렬, bucket을 정렬
# O(n), 최악의 경우 O(n²)

def bucket_sort(arr6):
    # Determine the range of the values
    max_val = ord('z') - ord('a') + 2  # 알파벳 범위 + 1(영어가 아닌 문자)
    
    # Create empty buckets
    buckets = [[] for _ in range(max_val)]
    
    # Distribute the elements into the appropriate buckets
    for word in arr6:
        if not word.isalpha():  # 알파벳 문자가 아닌 경우
            buckets[0].append(word)  # 영어가 아닌 문자용 버킷에 추가
        else:
            index = ord(word[0].lower()) - ord('a') + 1
            buckets[index].append(word.lower())
    
    # Sort the elements in each bucket and concatenate them
    sorted_arr = []
    for bucket in buckets:
        if len(bucket) > 1:  # 버킷에 하나 이상의 요소가 있는 경우에만 정렬 수행
            sorted_arr.extend(sorted(bucket))
        else:
            sorted_arr.extend(bucket)
    
    return sorted_arr

# Example usage
# arr6 = ["apple", "anana", "cherry", "date", "fig", "grape", "kiwi", "wikiwiki"]

arr6 = []

with open(name_path, 'r') as file:
    line = None
    while line != '':
        line = file.readline()
        arr6.append(line.strip('\n'))

Time = time.time()
sorted_arr = bucket_sort(arr6)
Total_Time6 = time.time() - Time

print("[Sorted array is]")
for i in range(len(sorted_arr)):
    print("[%04d] %s " %(i, sorted_arr[i]), end = "\n")

print(Total_Time6)
```

## 7. Radix Sort

```python
# 앞에서부터 각 자리수끼리 크기를 비교, 가장 긴 문자열의 길이를 기준으로 나머지 문자열의 길이를 0으로 채워서 맞춤
# O(kn), k는 자리수

def radix_sort(arr7):
    # Find the length of the longest string
    max_length = max(len(s) for s in arr7)
    
    # Pad all strings with zeros to have equal length
    padded_arr = [s.ljust(max_length, '\0') for s in arr7]
    
    # Sort the strings using radix sort
    for i in range(max_length - 1, -1, -1):
        # Create buckets for each character
        buckets = [[] for _ in range(256)]
        
        # Distribute the strings into the appropriate buckets
        for s in padded_arr:
            buckets[ord(s[i])].append(s)
        
        # Concatenate the strings in the buckets
        padded_arr = [s for bucket in buckets for s in bucket]
    
    # Remove padding and return sorted array
    sorted_arr = [s.replace('\0', '') for s in padded_arr]
    return sorted_arr

arr7 = []

with open(name_path, 'r') as file:
    line = None
    while line != '':
        line = file.readline()
        arr7.append(line.strip('\n'))

Time = time.time()
sorted_arr = radix_sort(arr7)
Total_Time7 = time.time() - Time

print("[Sorted array is]")
for i in range(len(sorted_arr)):
    print("[%04d] %s " %(i, sorted_arr[i]), end = "\n")

print(Total_Time7)
```

## 8. Tim Sort

```python
# Merge Sort와 Insertion Sort로 부터 기원한 하이브리드형 알고리즘
def merge(left, right):
    result = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result += left[i:]
    result += right[j:]
    return result

def timSort(arr8):
    if len(arr8) <= 1:
        return arr8
    min_run = 64
    n = len(arr8)
    runs = [arr8[i:i + min_run] for i in range(0, n, min_run)]
    for run in runs:
        insertion_sort(run)
    while len(runs) > 1:
        left, right = runs[:2]
        del runs[:2]
        merged = merge(left, right)
        runs.append(merged)
    return runs[0]

def insertion_sort(arr8):
    for i in range(1, len(arr8)):
        key_item = arr8[i]
        j = i - 1
        while j >= 0 and arr8[j] > key_item:
            arr8[j + 1] = arr8[j]
            j -= 1
        arr8[j + 1] = key_item

arr8 = []

with open(name_path, 'r') as file:
    line = None
    while line != '':
        line = file.readline()
        arr8.append(line.strip('\n'))


Time = time.time()
arr8 = timSort(arr8)
Total_Time8 = time.time() - Time

print("[Sorted array is]")
for i in range(len(arr8)):
    print("[%04d] %s " %(i, arr8[i]), end = "\n")

print(Total_Time8)
```
